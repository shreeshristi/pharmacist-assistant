```python
import cv2
import pytesseract
import matplotlib.pyplot as plt

# Set the path to Tesseract-OCR
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# Read the image
img = cv2.imread("medicine.png")

if img is None:
    print("Error: Image not found or path is incorrect")
else:
    # Convert BGR to RGB for correct display
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Convert image to grayscale (improves OCR accuracy)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Apply thresholding (removes background noise)
    thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

    # Display processed image
    plt.imshow(thresh, cmap='gray')
    plt.axis("off")
    plt.show()

    # Perform OCR (extract text)
    custom_config = r'--oem 3 --psm 6'  # Optimized OCR settings
    text = pytesseract.image_to_string(thresh, config=custom_config)

    print("Extracted Text:")
    print(text if text.strip() else "No text detected. Try improving the image quality.")
```
